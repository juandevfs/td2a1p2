Analisis de Ventas en Cajas Anual

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('historico.csv',sep=';', parse_dates=['Fecha'])
```

```python
df.head()
```

```python
df = df.set_index(pd.DatetimeIndex(df['Fecha']))
fig, ax = plt.subplots(figsize=(15,7))
df.groupby('Item').resample('m').sum()['Cajas'].unstack('Item', fill_value=0).plot(ax=ax)
```
